```python
nu_rel = 0.2  # relative bandwidth
eirp_vals = df["min_detectable_EIRP"].values

# Define log-spaced EIRP bins
log_eirp_min = np.log10(eirp_vals.min())
log_eirp_max = np.log10(eirp_vals.max())
log_bins = np.linspace(log_eirp_min, log_eirp_max, num=25)
eirp_bins = 10 ** log_bins

# Bin data
df['eirp_bin'] = pd.cut(df["min_detectable_EIRP"], bins=eirp_bins)
grouped = df.groupby('eirp_bin')

# Calculate values per bin
data = []
for bin_interval, group in grouped:
    Nstars = len(group)
    if Nstars == 0:
        continue
    mid_eirp = 10 ** ((np.log10(bin_interval.left) + np.log10(bin_interval.right)) / 2)
    transmitter_rate = 1 / (Nstars * nu_rel)
    # Error propagation
    N_err = np.sqrt(Nstars)
    T_err = N_err / (nu_rel * Nstars**2)
    data.append({
        "EIRP_mid": mid_eirp,
        "N*": Nstars,
        "Transmitter Rate": transmitter_rate,
        "Transmitter Rate Error": T_err
    })

# Create DataFrame
tr_df = pd.DataFrame(data)

# Plot with error bars
plt.figure(figsize=(8, 5))
plt.errorbar(
    np.log10(tr_df["EIRP_mid"]),
    np.log10(tr_df["Transmitter Rate"]),
    yerr=tr_df["Transmitter Rate Error"] / (tr_df["Transmitter Rate"] * np.log(10)),  # convert to log error
    fmt='o-',
    color='darkred',
    ecolor='gray',
    capsize=3
)
plt.xlabel("log(EIRP_min) [log Watts]")
plt.ylabel("log(Transmitter Rate)")
plt.title("log(Transmitter Rate) vs log(EIRP_min)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Filter stars within 1 kiloparsec (1000 parsecs)
# df_nearby = df[df['r_est'] <= 1]  # assuming r_est is in kpc

# Define ν_rel and constants
nu_rel = 0.2  # fractional bandwidth
log_eirp_min = np.log10(df['min_detectable_EIRP'].min())
log_eirp_max = np.log10(df['min_detectable_EIRP'].max())

# Use 15 log-spaced bins
log_bins = np.linspace(log_eirp_min, log_eirp_max, num=15)
eirp_bins = 10 ** log_bins

# Bin the EIRPmin values
df['eirp_bin'] = pd.cut(df['min_detectable_EIRP'], bins=eirp_bins)

# Group and compute transmitter rate and uncertainties
plot_data = []
for bin_interval, group in df.groupby('eirp_bin'):
    N = len(group)
    if N == 0:
        continue
    eirp_mid_log = np.log10(group['min_detectable_EIRP'].median())
    transmitter_rate = 1 / (N * nu_rel)
    # Error propagation: standard deviation of transmitter rate (Poisson assumption)
    transmitter_rate_err = 1 / (np.sqrt(N) * N * nu_rel)
    plot_data.append((eirp_mid_log, np.log10(transmitter_rate), transmitter_rate_err))

# Convert to arrays
eirp_log_vals, log_transmitter_rates, rate_errors = zip(*plot_data)

# Plot
plt.figure(figsize=(8, 5))
plt.errorbar(eirp_log_vals, log_transmitter_rates, 
             yerr=rate_errors, fmt='o-', capsize=3, label='Transmitter Rate')
plt.xlabel('log(EIRP$_{\\mathrm{min}}$) [W]')
plt.ylabel('log(Transmitter Rate)')
plt.title('Transmitter Rate vs EIRP$_{\\mathrm{min}}$ (15 Bins, d < 10 kpc)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

# CMD

```python
import numpy as np
from scipy.stats import gaussian_kde
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

def plot_density_colored_cmd(data_frame, x_col='bp_rp', y_col='abs_g_photogeo', 
                            figsize=(7, 9), point_size=10, cmap_name='plasma'):
    """
    Creates a density-colored Color-Magnitude Diagram (CMD) from stellar data.
    
    Args:
        data_frame (pd.DataFrame): DataFrame containing stellar data
        x_col (str): Column name for color index (x-axis)
        y_col (str): Column name for absolute magnitude (y-axis)
        figsize (tuple): Figure dimensions (width, height) in inches
        point_size (int): Size of scatter plot points
        cmap_name (str): Matplotlib colormap name
    
    Returns:
        matplotlib.figure.Figure: The created figure object
    """
    # Extract color and magnitude data
    x = data_frame[x_col]
    y = data_frame[y_col]
    
    # Remove any rows with NaNs for clean plotting
    mask = x.notna() & y.notna()
    x_clean = x[mask]
    y_clean = y[mask]
    
    # Compute point density
    xy = np.vstack([x_clean, y_clean])
    density = gaussian_kde(xy)(xy)
    sqrt_density = np.sqrt(density)  # Square root for better visual contrast
    
    # Create figure and plot
    fig = plt.figure(figsize=figsize)
    
    # Create a proper colormap and normalization
    cmap = plt.cm.get_cmap(cmap_name)
    norm = mcolors.Normalize(vmin=sqrt_density.min(), vmax=sqrt_density.max())
    
    # Use the scalar mappable for consistent coloring
    scatter = plt.scatter(x_clean, y_clean, c=sqrt_density, s=point_size, 
                         cmap=cmap, norm=norm, edgecolor='none')
    
    # Format plot
    plt.gca().invert_yaxis()  # Astronomical convention: brighter stars at top
    plt.xlabel('BP − RP (mag)')
    plt.ylabel('Absolute G-band Magnitude')
    plt.title('Color-Magnitude Diagram (Density Colored) Bailer-Jones')
    plt.colorbar(scatter, label='√(Relative Density)')
    plt.grid(True)
    plt.tight_layout()
    
    return fig

# Usage example:
fig_1 = plot_density_colored_cmd(region_1_1)
plt.show(fig_1)
```

# Angular Separation Function

```python
import numpy as np
from astropy.coordinates import SkyCoord
import astropy.units as u

def angular_separation_arcmin(ra1, dec1, ra2, dec2):
    """
    Compute angular separation between two points (RA/Dec, degrees) in arcminutes.
    All inputs and output are floats.
    """
    c1 = SkyCoord(ra1, dec1, unit='deg')
    c2 = SkyCoord(ra2, dec2, unit='deg')
    sep = c1.separation(c2)
    return sep.arcminute  # Output: float, arcminutes
```

# Beam Scaling Function

```python
def beam_scaling(beam_ra, beam_dec, star_ra, star_dec, fwhm_arcmin):
    """
    Calculate angular separation and Gaussian beam scaling factor for a star.
    All coordinates in degrees, FWHM and output in arcminutes.
    Returns: (theta_arcmin: angular separation between star and beam center, scaling_factor: normalized sensitivity of the star, 
    eirp_boost: how many times larger the EIRPmin of the star must be than the beam's center)
    """
    theta_arcmin = angular_separation_arcmin(beam_ra, beam_dec, star_ra, star_dec)
    exponent = -4 * np.log(2) * (theta_arcmin / fwhm_arcmin)**2
    scaling_factor = np.exp(exponent)
    eirp_boost = 1.0 / scaling_factor if scaling_factor > 0 else np.inf
    return theta_arcmin, scaling_factor, eirp_boost
```

```python
import pandas as pd
import numpy as np
from astropy.coordinates import SkyCoord
import astropy.units as u

def process_beam(beam_ra, beam_dec, fwhm_arcmin, catalog_df):
    """
    Add beam sensitivity columns to existing DataFrame
    Inputs:
        beam_ra, beam_dec: Beam center (degrees)
        fwhm_arcmin: Beam width (arcminutes)
        catalog_df: DataFrame containing 'source_id', 'ra', 'dec'
    Returns: Modified DataFrame with new columns
    """
    # Pre-allocate arrays for performance
    n = len(catalog_df)
    theta_arr = np.empty(n)
    scaling_arr = np.empty(n)
    eirp_arr = np.empty(n)
    
    # Compute values for all stars
    for i, (_, row) in enumerate(catalog_df.iterrows()):
        theta_arr[i], scaling_arr[i], eirp_arr[i] = beam_scaling(
        beam_ra, beam_dec, row['ra'], row['dec'], fwhm_arcmin
        )
    
    # Add new columns to original DataFrame
    catalog_df = catalog_df.copy()  # Avoid SettingWithCopyWarning
    catalog_df['theta_arcmin'] = theta_arr
    catalog_df['scaling_factor'] = scaling_arr
    catalog_df['eirp_boost'] = eirp_arr
    
    return catalog_df
```
